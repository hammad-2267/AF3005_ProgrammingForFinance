## **ğŸŸ¢ Part 1: Loan Eligibility & Interest Rate Calculation [4 marks]**  

ğŸ“Œ **Requirement:**  
SecureBank only approves loans if the **customer meets all eligibility criteria**. The system should:  
âœ” **Check if the customer is employed.**  
âœ” **Verify if the income is at least PKR 50,000.**  
âœ” **Check credit score:**  
   - **750+ â†’ 5% Interest Rate**  
   - **650 - 749 â†’ 8% Interest Rate**  
   - **Below 650 â†’ Loan Rejected**  
âœ” **If the applicant is unemployed, the loan is rejected immediately.**  

### **Implementation Steps:**  
ğŸ”¹ **Step 1:** Structure the `if-elif-else` statements for this logic.  
ğŸ”¹ **Step 2:** Implement a Python program using **ipywidgets** for interactive user input (income, credit score, employment status) to decide whether the loan is approved or rejected.

---

```python
import ipywidgets as widgets

def loan_status(employment_status, income, credit_score):
    if employment_status == "Unemployed":
        return "Loan Rejected: Unemployed."
    if income < 50000:
        return "Loan Rejected: Income < PKR 50,000."
    if credit_score >= 750:
        return "Loan Approved! Interest Rate: 5%"
    if 650 <= credit_score < 750:
        return "Loan Approved! Interest Rate: 8%"
    return "Loan Rejected: Credit score < 650."

employment_status = widgets.RadioButtons(options=["Employed", "Unemployed"], description="Employment:")
income = widgets.IntSlider(min=0, max=200000, step=1000, description="Income (PKR):")
credit_score = widgets.IntSlider(min=300, max=850, step=1, description="Credit Score:")
check_button = widgets.Button(description="Check Eligibility")

def on_check_eligibility(_):
    result = loan_status(employment_status.value, income.value, credit_score.value)
    print("\n--- Result ---")
    print(result)

check_button.on_click(on_check_eligibility)
widgets.VBox([employment_status, income, credit_score, check_button])
```

## **ğŸŸ¢ Part 2: Investment Risk Assessment [4 marks]**  

ğŸ“Œ **Requirement:**  
SecureBank offers **investment analysis services**. The system should evaluate a **portfolio of stocks** and classify the risk level based on stock returns:  
âœ” If **any stock has a negative return**, the portfolio is **High Risk**.  
âœ” If **all stocks have positive returns**, but at least one return is below 5%, classify as **Medium Risk**.  
âœ” If **all stock returns are 5% or above**, classify as **Low Risk**.  

### **Implementation Steps:**  
ğŸ”¹ **Step 1:** Use loops to iterate through the list of stock returns.  
ğŸ”¹ **Step 2:** Implement `if-elif` conditions to classify the risk.  
ğŸ”¹ **Step 3:** Write a Python program using **ipywidgets** to allow users to **input stock returns interactively** and receive a risk assessment.

---

```python
import ipywidgets as widgets

def determine_risk_level(stock_returns):
    if any(return_value < 0 for return_value in stock_returns):
        return "High Risk"
    if all(return_value >= 5 for return_value in stock_returns):
        return "Low Risk"
    return "Medium Risk"

stock_returns_input = widgets.Text(
    placeholder="Enter returns (e.g., 6, -2, 10)",
    description="Returns:"
)
result_display = widgets.Output()
risk_button = widgets.Button(description="Determine Risk")

def handle_risk_check(_):
    try:

        stock_returns = [float(value.strip()) for value in stock_returns_input.value.split(",")]
        result_display.clear_output()
        with result_display:
            print(f"Risk: {determine_risk_level(stock_returns)}")
    except ValueError:
        result_display.clear_output()
        with result_display:
            print("Invalid input. Use numbers separated by commas.")

risk_button.on_click(handle_risk_check)
widgets.VBox([stock_returns_input, risk_button, result_display])
```

## **ğŸŸ¢ Part 3: Loan Repayment Tracker [4 marks]**  

ğŸ“Œ **Requirement:**  
Customers who receive a loan should be able to track their **loan balance** as they make monthly payments. The system should:  
âœ” Start with an **initial loan balance** (e.g., PKR 500,000).  
âœ” Deduct a **fixed monthly payment** (e.g., PKR 25,000).  
âœ” Continue tracking until **loan balance reaches zero**.  
âœ” Display the remaining balance **after each payment**.  

### **Implementation Steps:**  
ğŸ”¹ **Step 1:** Choose an appropriate loop (`for` or `while`).  
ğŸ”¹ **Step 2:** Ensure the loop **stops once the loan is fully paid**.  
ğŸ”¹ **Step 3:** Implement a Python program using **ipywidgets** to **simulate loan repayment interactively**.

---

```python
import ipywidgets as widgets

def repayment_schedule(balance, payment):
    month = 1
    while balance > 0:
        balance = max(balance - payment, 0)
        print(f"Month {month}: Balance = PKR {balance:,.2f}")
        month += 1

balance_slider = widgets.IntSlider(min=100000, max=1000000, step=10000, value=500000, description="Balance:")
payment_slider = widgets.IntSlider(min=1000, max=50000, step=1000, value=25000, description="Payment:")
calculate_button = widgets.Button(description="Calculate Repayment")

def handle_button_click(_):
    print("\n--- Repayment Schedule ---")
    repayment_schedule(balance_slider.value, payment_slider.value)

calculate_button.on_click(handle_button_click)
widgets.VBox([balance_slider, payment_slider, calculate_button])
```


## **ğŸŸ¢ Part 4: Stock Price Monitoring and Trading Strategy [4 marks]**  

ğŸ“Œ **Requirement:**  
A stock trader wants to **track stock prices daily** and **sell when the price reaches PKR 200**. The system should:  
âœ” Iterate through a **list of stock prices**.  
âœ” **Skip missing stock data** (`None` values).  
âœ” Stop tracking **once the price reaches PKR 200**.  

### **Implementation Steps:**  
ğŸ”¹ **Step 1:** Handle missing stock data using `continue`.  
ğŸ”¹ **Step 2:** Stop tracking once the stock hits the target price (`break`).  
ğŸ”¹ **Step 3:** Write a Python program using **ipywidgets** to **process stock prices interactively** and trigger alerts when conditions are met.

---

```python
def track_stock_prices(stock_prices, sell_price):
    for index, stock_price in enumerate(stock_prices, 1):
        if stock_price is None:
            print(f"Day {index}: Missing data (skipped)")
            continue
        if stock_price >= sell_price:
            print(f"Day {index}: Price = PKR {stock_price:.2f} (Target reached! Sell stock.)")
            break
        print(f"Day {index}: Price = PKR {stock_price:.2f}")

prices = [150, None, 180, 200, 210]
target_price = 200

print("--- Stock Price Tracker ---")
track_stock_prices(prices, target_price)
```


## **ğŸŸ¢ Part 5: Currency Exchange Rate Tracker [4 marks]**  

ğŸ“Œ **Requirement:**  
SecureBank provides **real-time currency exchange tracking**. The system should:  
âœ” Start at **PKR 290/USD**.  
âœ” Increase by **1 PKR per day** until it reaches **PKR 300/USD**.  
âœ” Print exchange rates daily and stop when the **target rate is reached**.  

### **Implementation Steps:**  
ğŸ”¹ **Step 1:** Choose a suitable loop (`for` or `while`).  
ğŸ”¹ **Step 2:** Stop the loop when the exchange rate reaches the target.  
ğŸ”¹ **Step 3:** Implement a Python program using **ipywidgets** to **track the currency exchange rate interactively**.

---

```python
def monitor_exchange_rate(initial_rate, goal_rate):
    rate = initial_rate
    day_count = 1
    while rate <= goal_rate:
        print(f"Day {day_count}: Exchange Rate = PKR {rate}/USD")
        if rate == goal_rate:
            print("Target rate reached! Stopping tracker.")
            break
        rate += 1
        day_count += 1

start_rate = 290
target_rate = 300

print("--- Currency Exchange Rate Monitor ---")
monitor_exchange_rate(start_rate, target_rate)
```
